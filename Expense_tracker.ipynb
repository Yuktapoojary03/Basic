```python
import os

expenses = []

file_path = "/content/expense.csv"

# Load existing data
if os.path.exists(file_path):
    file = open(file_path, "r")
    next(file)  # skip header
    for line in file:
        data = line.strip().split(",")
        if len(data) == 4:
            expenses.append({
                "date": data[0],
                "category": data[1],
                "amount": float(data[2]),
                "description": data[3]
            })
    file.close()

while True:
    print("\n--- Expense Tracker ---")
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. Total Expense")
    print("4. Filter by Category")
    print("5. Exit")

    choice = input("Enter choice: ")

    # Add Expense
    if choice == "1":
        date = input("Enter date (YYYY-MM-DD): ")
        category = input("Enter category: ")
        amount = float(input("Enter amount: "))
        desc = input("Enter description: ")

        expenses.append({
            "date": date,
            "category": category,
            "amount": amount,
            "description": desc
        })

        print("Expense added!")

    # View Expenses
    elif choice == "2":
        if len(expenses) == 0:
            print("No expenses found.")
        else:
            for e in expenses:
                print(e)

    # Total Expense
    elif choice == "3":
        total = 0
        for e in expenses:
            total += e["amount"]
        print("Total Expense:", total)

    # Filter by Category
    elif choice == "4":
        cat = input("Enter category: ")
        found = False
        for e in expenses:
            if e["category"].lower() == cat.lower():
                print(e)
                found = True
        if not found:
            print("No expenses in this category.")

    # Exit & Save
    elif choice == "5":
        file = open(file_path, "w")
        file.write("date,category,amount,description\n")
        for e in expenses:
            file.write(f"{e['date']},{e['category']},{e['amount']},{e['description']}\n")
        file.close()

        print("Data saved. Exiting...")
        break

    else:
        print("Invalid choice.")
```
