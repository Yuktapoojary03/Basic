```python
import os

accounts = []
file_path = "/bank_data.csv"

# Load existing data
if os.path.exists(file_path):
    file = open(file_path, "r")
    for line in file:
        data = line.strip().split(",")
        if len(data) == 4:
            accounts.append({
                "acc_no": data[0],
                "name": data[1],
                "password": data[2],
                "balance": float(data[3])
            })
    file.close()

while True:
    print("\n--- Online Banking System ---")
    print("1. Create Account")
    print("2. Login")
    print("3. Exit")

    choice = input("Enter choice: ")

    # Create Account
    if choice == "1":
        acc_no = input("Enter Account Number: ")
        name = input("Enter Name: ")
        password = input("Set Password: ")
        balance = float(input("Enter Initial Balance: "))

        accounts.append({
            "acc_no": acc_no,
            "name": name,
            "password": password,
            "balance": balance
        })

        print("Account created successfully!")

    # Login
    elif choice == "2":
        acc_no = input("Enter Account Number: ")
        password = input("Enter Password: ")

        user = None
        for acc in accounts:
            if acc["acc_no"] == acc_no and acc["password"] == password:
                user = acc
                break

        if user is None:
            print("Invalid login!")
        else:
            print("Login successful!")

            # User Menu
            while True:
                print("\n1. Deposit")
                print("2. Withdraw")
                print("3. Check Balance")
                print("4. Logout")

                ch = input("Enter choice: ")

                # Deposit
                if ch == "1":
                    amount = float(input("Enter amount: "))
                    user["balance"] += amount
                    print("Deposited successfully!")

                # Withdraw
                elif ch == "2":
                    amount = float(input("Enter amount: "))
                    if amount <= user["balance"]:
                        user["balance"] -= amount
                        print("Withdrawal successful!")
                    else:
                        print("Insufficient balance!")

                # Check Balance
                elif ch == "3":
                    print("Current Balance:", user["balance"])

                # Logout
                elif ch == "4":
                    break

                else:
                    print("Invalid choice!")

    # Exit and Save
    elif choice == "3":
        file = open(file_path, "w")
        for acc in accounts:
            file.write(f"{acc['acc_no']},{acc['name']},{acc['password']},{acc['balance']}\n")
        file.close()

        print("Data saved. Exiting...")
        break

    else:
        print("Invalid choice!")
```
