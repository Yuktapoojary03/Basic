```python
import os

students = []

# Load data from file if exists
if os.path.exists("/students.csv"):
    file = open("/students.csv", "r")
    for line in file:
        data = line.strip().split(",")
        if len(data) == 3:
            students.append({"id": data[0], "name": data[1], "age": data[2]})
    file.close()

while True:
    print("\n--- Student Management System ---")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Delete Student")
    print("6. Exit")

    choice = input("Enter your choice: ")

    # Add Student
    if choice == "1":
        sid = input("Enter ID: ")
        name = input("Enter Name: ")
        age = input("Enter Age: ")

        students.append({"id": sid, "name": name, "age": age})
        print("Student added successfully!")

    # View Students
    elif choice == "2":
        if len(students) == 0:
            print("No students found.")
        else:
            for s in students:
                print(s)

    # Search Student
    elif choice == "3":
        sid = input("Enter ID to search: ")
        found = False
        for s in students:
            if s["id"] == sid:
                print("Found:", s)
                found = True
                break
        if not found:
            print("Student not found.")

    # Update Student
    elif choice == "4":
        sid = input("Enter ID to update: ")
        found = False
        for s in students:
            if s["id"] == sid:
                s["name"] = input("Enter new name: ")
                s["age"] = input("Enter new age: ")
                print("Student updated!")
                found = True
                break
        if not found:
            print("Student not found.")

    # Delete Student
    elif choice == "5":
        sid = input("Enter ID to delete: ")
        found = False
        for s in students:
            if s["id"] == sid:
                students.remove(s)
                print("Student deleted!")
                found = True
                break
        if not found:
            print("Student not found.")

    # Exit
    elif choice == "6":
        # Save data to file
        file = open("students.txt", "w")
        for s in students:
            file.write(s["id"] + "," + s["name"] + "," + s["age"] + "\n")
        file.close()

        print("Data saved. Exiting...")
        break

    else:
        print("Invalid choice. Try again.")
```
